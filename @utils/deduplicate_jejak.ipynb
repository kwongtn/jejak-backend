```python
import hashlib
import os
```

```python
unique_hashes = set()
counter = 0

INPUT_FILENAME = "test.json"
OUTPUT_FILENAME = "test_dedup.json"

INPUT_FILESIZE = os.path.getsize(INPUT_FILENAME)

with open(INPUT_FILENAME, "r") as in_file:
    with open(OUTPUT_FILENAME, "a") as out_file:

        while True:
            line = in_file.readline()
            if not line:
                break

            hash = hashlib.sha512(line.encode("utf-8")).hexdigest()

            if hash not in unique_hashes:
                unique_hashes.add(hash)
                out_file.write(line)

            counter += 1

            if counter % 10000 == 0:
                print(".", end="")

            if counter % 1000000 == 0:
                print("")

OUTPUT_FILESIZE = os.path.getsize(OUTPUT_FILENAME)

print("")
print("Input filesize: ", INPUT_FILESIZE)
print("Output filesize: ", OUTPUT_FILESIZE)
print("Filesize reduction: ", OUTPUT_FILESIZE * 100 / INPUT_FILESIZE, "%")
```
